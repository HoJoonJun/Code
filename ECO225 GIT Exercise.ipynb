```python
#First, code that returns the most frequently appearing companies.


def freq_name(data_frame, column_name, n: int):
    """Given a dataframe and a column name, return a Series of each company along with the number of times they appear.
    n specfies up to the n-th most common.
    This function will be used for all the column's most frequent entries are.
    """
    # Checks whether the column name is in the df
    if column_name not in data_frame.columns:
        print('Invalid Column')
        return None
    
    # uses the value_counts() function on df
    entry_counts = data_frame[column_name].value_counts()
    
    
    return entry_counts.head(n)
```

```python
#Removing rows with a salary greater than 7-digits, as these are going to be massive outliers
def remove_above_threshold(df, column_name, n):
    """
    """
    mask = df[column_name] <= n
    new_df = df[mask]
    return new_df
remove_above_threshold(dfnew, 'annual_base_pay', 1000000)
```

```python
#Next, average Salary and max salary

def average_col(data_frame, column_name):
    """
    
    """
    # Checks whether the column name is in the df
    if column_name not in data_frame.columns:
        print('Invalid Column')
        return None
    
    # uses the value_counts() function on df
    avg = dfnew[column_name].sum() /len(dfnew[column_name])
    
    #note this rounds the salary to the nearest whole number.
    return avg.round(0)
average_col(dfnew, 'annual_base_pay')
```
